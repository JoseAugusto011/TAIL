```python
def VerificarValores(df):
    
    #Ver quais os valores únicos de cada coluna
    #Vericar qtd de repetições das variaveis em cada coluna 
    print("Informações sobre os dados: \n")
    for col in df.columns:
        print(df[f"{col}"].value_counts())
        print("\n")
        
    print("\n")
    print(df.info())
    print("\n")
    print(f'Há duplicatas? {df.duplicated().values.any()}')
    print(f'Há valores NaN? {df.isna().values.any()}')
    
    
```

```python
VerificarValores(mr1)
```

```python
VerificarValores(mr2)
```
